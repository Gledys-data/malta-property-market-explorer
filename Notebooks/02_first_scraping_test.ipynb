## Week 1 - Day 2

### Objective

Test whether Python can access a Malta property listings page and read its HTML content.

Website tested:

https://www.propertymarket.com.mt/for-sale/?locations%5B%5D=Birkirkara

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date

url = "https://www.propertymarket.com.mt/for-sale/?locations%5B%5D=Birkirkara"

response = requests.get(url)

print("Status code:", response.status_code)
print("Page length:", len(response.text))
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")

print(soup.title)
```

```python
print(soup.prettify()[:3000])
```

```python
text = soup.get_text()

print(text.find("Birkirkara"))
```

```python
print(soup.title.text)
```

```python
print(len(soup.find_all("a")))
```

```python
print(soup.title.text)

print(len(soup.find_all("a")))
```

```python
links = soup.find_all("a")

for link in links[:20]:
    print(link.get("href"))
```

```python
property_links = []

for link in links:
    href = link.get("href")

    if href and "/view/" in href:
        property_links.append(href)

print("Properties found:", len(property_links))

print("\nFirst property:")
print(property_links[0])
```

```python
first_url = property_links[0]

property_response = requests.get(first_url)

print("Status:", property_response.status_code)
print("Length:", len(property_response.text))
```

```python
print(property_response.text)
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}

property_response = requests.get(
    first_url,
    headers=headers
)

print("Status:", property_response.status_code)
print("Length:", len(property_response.text))
```

```python
print(property_response.text)
```

```python
print(first_url)
```

```python
print(property_response.url)
```

```python
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/137.0.0.0 Safari/537.36"
    )
}

property_response = requests.get(
    first_url,
    headers=headers
)

print("Status:", property_response.status_code)
print("Final URL:", property_response.url)
print("Length:", len(property_response.text))
```

```python
property_soup = BeautifulSoup(
    property_response.text,
    "html.parser"
)

print(property_soup.title.text)
```

```python
page_text = property_soup.get_text()

print(page_text[:5000])
```

```python
price_matches = property_soup.find_all(string=lambda text: text and "€" in text)

for p in price_matches[:20]:
    print(p.strip())
```
