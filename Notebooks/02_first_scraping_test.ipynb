## 1. Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date
```

## 2. Define the Target URL

The project begins by selecting a property listing page from PropertyMarket.com.mt.

For this notebook, the selected location is **Birkirkara**.

The goal is to understand the webpage structure before attempting to extract property information.

```python
target_url= "https://www.propertymarket.com.mt/for-sale/?locations%5B%5D=Birkirkara"
```

```python
listing_response = requests.get(target_url)

print("Status code:", listing_response.status_code)
print("Page length:", len(listing_response.text))
```

## 3. Handle the 403 Forbidden Response

The first request returned a **403 Forbidden** status code.

This means the website understood the request, but refused access. This often happens because the request does not look like it comes from a normal web browser.

To test this, I added a `User-Agent` header. This header identifies the request as coming from a browser.

```python
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/137.0.0.0 Safari/537.36"
    )
}

listing_response = requests.get(target_url, headers=headers)

print("Status code:", listing_response.status_code)
print("Final URL:", listing_response.url)
print("Page length:", len(listing_response.text))
```

## 4. Parse the HTML with BeautifulSoup

After receiving the webpage response, the HTML content can be parsed using BeautifulSoup.

This creates a `soup` object, which allows Python to search and inspect the HTML structure.

```python
soup = BeautifulSoup(listing_response.text, "html.parser")

print(soup.title.text)
```

## 5. Inspect the HTML Structure

Before extracting property information, it is useful to inspect the HTML structure of the webpage.

This helps identify the HTML elements that contain property links, titles, prices, and other useful information.

```python
print(soup.prettify()[:1000])
```

```python
print(soup.title.text)
```

```python
page_text = soup.get_text()

print(page_text[:1000])
```

```python
keyword = "Birkirkara"

print(f'"{keyword}" found at position:', page_text.find(keyword))
```

## 6. Explore the Hyperlinks

Most property websites use hyperlinks (`<a>` tags) to connect the listing page with individual property pages.

The first step is to inspect these links and understand their structure before extracting property URLs.

```python
links = soup.find_all("a")

print("Total links found:", len(links))
```

```python
for link in links[:20]:
    print(link.get("href"))
```

## 7. Extract Property Links

After inspecting the hyperlinks, it became clear that every property listing contains the pattern `/view/` in its URL.

Using this observation, the property links can be filtered from all other navigation links on the page.

```python
property_links = []

for link in links:
    href = link.get("href")

    if href and "/view/" in href:
        property_links.append(href)

print("Property links found:", len(property_links))
```

```python
print("First five property links:")

for property_url in property_links[:5]:
    print(property_url)
```

### Observation

The extracted property URLs contain the location name "Birkirkara", which matches the location specified in the original search URL.

This suggests that the filtering on the website is working correctly. In later notebooks, the property location will be verified directly from each property's webpage.

## 8. Remove Duplicate Property Links

Some property listings appear multiple times on the webpage.

To avoid collecting duplicate records, duplicate URLs are removed before continuing with the analysis.

```python
unique_property_links = list(dict.fromkeys(property_links))

print("Unique property links:", len(unique_property_links))
```

```python
print("First five unique property links:")

for property_url in unique_property_links[:5]:
    print(property_url)
```

### Observation

The first five unique property links show different property listings, confirming that duplicate URLs were successfully removed.

Removing duplicate links helps ensure that each property is processed only once when building the dataset.

## 9. Open the First Property Page

After extracting the property URLs, the next step is to visit an individual property page.

This allows us to inspect its HTML structure and identify where important information such as the title, price, property type, bedrooms, and bathrooms is stored.

```python
first_property_url = unique_property_links[0]

print(first_property_url)
```

```python
property_response = requests.get(
    first_property_url,
    headers=headers
)

print("Status code:", property_response.status_code)
print("Page length:", len(property_response.text))
```

```python
property_soup = BeautifulSoup(
    property_response.text,
    "html.parser"
)
```

```python
print(property_soup.title.text)
```

## 10. Inspect the Property Page

The property page was successfully retrieved.

Before extracting specific information, it is useful to inspect the HTML and visible text to understand where the property details are located.

```python
print(property_soup.prettify()[:1500])
```

```python
property_text = property_soup.get_text()

print(property_text[:3000])
```

```python
price_matches = property_soup.find_all(
    string=lambda text: text and "€" in text
)

for price in price_matches[:10]:
    print(price.strip())
```

### Observation

Searching for the euro symbol successfully identified several price-related values on the property page.

However, not every value corresponds to the property's sale price. Some values belong to mortgage calculators, advertisements, or other page elements.

This highlights the importance of locating the correct HTML element before extracting data automatically.
